# Light diffraction at the slit and grating

The following experiment was carried out as part of a laboratory internship at the University of Regensburg.
No precise information about the task is given and therefore only the resulting measurement data is shown and evaluated.



When carrying out this experiment, various diffraction images of single and double slits as well as of a grating are recorded. 
A He-Ne laser with a wavelength of $λ = 632.8nm$ as well as a linear shifter and a phototransistor as well as an analog-digital converter are used
to record the measured diffraction images.

The laser used is clamped on the linear slider at a distance of $x_L = 90mm$ from the zero point. The detector is located at $x_D = 1970mm$.
The computer controls a stepper motor that can move the light detector. The total distance is 1175 steps or channels. 
Each channel corresponds to a distance of $12.5µm$. These are used as the $x$-axis in the intensity distribution. 
During the measurement, each step has a measured voltage value that is proportional to the incident intensity.

## Light diffraction at the single slit

For the first recording of the intensity curve, a single slit with a slit width of $b = 0.2mm$ is used. 
The gap is clamped at a distance of $x_E = 1605mm$ from the zero point on the linear slider.

```python
data1 = pd.read_csv('Einfachspalt_1.csv')

data1.plot()
plt.title("Intensity distribution at the single slit")
plt.xlabel("Number of Channels")
plt.ylabel("Intensity I")
plt.show()
```

### Width of the single slit

To eveluate the data further a calculation of the width of the single slit is needed to verify the accuracy. <br>
The distance s between the intensity maximum of 0th order and 1st order is 148 channels and therefore $s = 148 \cdot 12,5µm = 1850µm = 1.85mm$. <br>
The distance $\Delta x$ between the measuring device and the single slit is $\Delta x = x_D − x_E = 1970mm − 1605mm = 365mm = 0.365m.$ <br>
This follows for the width of the single gap 

```python
%%latex
\begin{align*}
\sin (\Phi_k , max) = \frac{\lambda \cdot k}{d} \enspace \Rightarrow \enspace d = \frac{\lambda \cdot k \cdot \Delta x}{s} = \frac{632,8 \cdot 10^{-6}mm \cdot 1 \cdot 365 mm}{1,85 mm} = 0,12 mm
\end{align*}
```

The calculated value lies in the measurement accuracy of the specified gap width of $b = 0.2mm$ <br>
The deviation can be attributed to the measurement inaccuracy of the measuring device used or to inaccuracies when reading and adjusting the single gap.

### Comparison with theory

The intensity distribution for a single slit can be calculated using the formula: 

```python
%%latex
\begin{align*}
    I(\Phi) = I_0 \left\lbrace \frac{\sin \frac{\Phi}{2}}{\frac{\Phi}{2}} \right\rbrace^2 \text{ with } \Phi = \frac{2 \pi}{\lambda} \cdot b \cdot \sin \Phi \text{ and } I_0 = I(\Phi = 0)
\end{align*}
```

Here $\Phi$ is the diffraction angle, i.e. the angle by which the light scattered by the slit deviates from the original beam direction. $\Phi$ is the opening angle of the gap and depends on the wavelength $\lambda$ of the scattered radiation and the width $b$ of the gap. $I_0$ is the intensity of undiffracted light, i.e. the intensity observed when $\Phi = 0$. <br>
The formula for the intensity distribution can be derived as follows: <br>
First we look at the interference pattern that results from diffraction at a single slit. Interference occurs at the diffraction point from the waves emanating from both sides of the gap. This interference is described by the Huygens-Fresnel principle. <br>
The Huygens-Fresnel principle states that every point on the wavefront of an electromagnetic wave emits a spherical wave called secondary Huygen waves. <br>
The Huygen secondary waves emanating from both sides of the gap overlap and form an interference pattern. The intensity of the light depends on the square root of the amplitude, which is why the interference pattern is reflected in the intensity of the light. <br>
The interference pattern can be described using the interference condition, which states that the phase difference between the two waves coming together at the diffraction point must be a multiple of 2π in order to achieve maximum intensity. The phase difference depends on the wavelength $\lambda$, the distance $\Delta x$ between the maxima and the angle $\Phi$ through which the light is deflected. This results in the distance $\Delta x$, the path difference $\Delta s$ and the phase difference $\Delta \Phi$:

```python
%%latex
\begin{align*}
    \Delta x &= d \cdot \sin \Phi \\
    \Delta s &= \sqrt{d^2 + (d \sin \Phi)^2} - d \approx d \cdot \Phi \\
    \Delta \Phi &= \frac{2 \pi}{\lambda} \cdot \Delta s = \frac{2 \pi}{\lambda} \cdot d \sin \Phi
\end{align*}
```

The division of the light beam into $N$ individual beams at a distance $q = \frac{b}{N}$ leads to each individual resulting in a phase difference of $\Delta \Phi$. So we can describe the entire $\vec{E}$ field in one point:

```python
%%latex
\begin{align*}
E & = A \sum\limits_{j=1}^N e^{i(\omega t + \phi_j)}
& = Ae^{i\omega t} \sum\limits_{j=1}^N e^{i(j-1)\Delta \phi}
& = Ae^{i\omega t} \frac{e^{iN\phi}-1}{e^{i\Delta \phi}-1}
& = Ae^{i\omega t} e^{i\frac{N-1}{2}\Delta \phi}\left(\frac{\sin\left(\frac{N}{2}\Delta \phi\right)}{\sin \left(\frac{\Delta \phi}{2} \right)} \right)
\end{align*}
```
